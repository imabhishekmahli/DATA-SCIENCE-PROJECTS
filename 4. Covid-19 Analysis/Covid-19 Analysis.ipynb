# Q 1. Show the no. of confirmed,Death and Cured cases in each State/UT ?

```python
covid.head(2)
```

```python
covid.groupby('State/UnionTerritory').sum()
```

```python
## confirmed cases in Ascending order
covid.groupby('State/UnionTerritory')['Confirmed'].sum().sort_values(ascending=True)   
```

```python
## group by state/UT and their confirmed and cured cases
covid.groupby('State/UnionTerritory')['Confirmed','Cured'].sum()
```

# Q 3. In which State/UT max no. confirmed cases were recorded ?

```python
covid.head(2)
```

```python
covid.rename(columns={'State/UnionTerritory':'Region'},inplace=True)
```

```python
covid.head(2)
```

```python
covid.groupby('Region')['Confirmed'].sum().sort_values(ascending=False).head()  # solved
```
